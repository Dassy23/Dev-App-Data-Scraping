```python
# coding=utf-8
#import os
from selenium import webdriver
import time
import pandas as pd
import json
import re

total_df = pd.DataFrame()
#get weblinks for json file
with open('development_applications.json', encoding='utf-8') as data_file:
    data = json.loads(data_file.read())
features = data["features"]
prop_list=[]
for key in features:
    prop= key["properties"]
    prop_list.append(prop)
weblinks = []
for key in prop_list:
    weblink = key["WEBLINK"]
    weblinks.append(weblink)

```

```python
my_df = pd.DataFrame(weblinks)
my_df.columns = ['weblinks']
my_df.set_index('weblinks')
#my_df.to_csv('weblinks.csv', index=False, header=False)
my_df['weblinks'] = my_df['weblinks'].astype('str')
my_df.dtypes
```
